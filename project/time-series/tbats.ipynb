```python

import pandas as pd
from tbats import TBATS
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('Weather_Data_2020_2024.csv', sep=',', decimal='.', index_col=0, parse_dates=True)

series = df['temperature']
```

```python
df['temperature'].plot(title='temperature over time')
```

```python
estimator = TBATS(seasonal_periods=[24, 7*24])
model = estimator.fit(series)

```

```python
y_forecasted, confidence_intervals = model.forecast(steps=24)
```

```python
plt.figure(figsize=(15,5))
plt.plot(series, label='Historical')
plt.plot(pd.date_range(series.index[-1], periods=96, freq='15min'), y_forecasted, label='Forecast')
plt.legend()
```
